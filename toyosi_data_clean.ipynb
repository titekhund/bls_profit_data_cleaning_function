```python
import pandas as pd

def toyosi_function(file_path):
    data = pd.read_csv(file_path)

    dates = pd.date_range(start="1947-01-01", periods=4*77, freq='Q-JAN')

    reshaped_data = pd.DataFrame(data.values[1:, 2:], index=data.iloc[1:, 1], columns=dates).T
    reshaped_data = reshaped_data.apply(pd.to_numeric, errors='coerce')
    reshaped_data.dropna(axis=1, how='all', inplace=True)

    return reshaped_data


```

```python

clean_data = toyosi_function("/content/Figure_1.csv")
clean_data.head()
```
